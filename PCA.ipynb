```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
faces = np.loadtxt("faces.csv", delimiter=',')
example_image = faces[0]
plt.imshow(example_image.reshape((64, 64)))
```

a)

```python
Z = faces - faces.mean(axis=0)
C = 1 / len(faces) * Z.T @ Z
```

```python
eigh = np.linalg.eigh(C)
eigenvalues = eigh[0]
eigenvectors = eigh[1]
```

```python
eyeglasses_detector = eigenvectors[:, -7]
```

```python
plt.imshow(eyeglasses_detector.reshape((64,64)))
```

b)

```python
output = abs(faces @ eyeglasses_detector)
```

```python
top20 = np.argpartition(output, -20)[-20:] #idxmax of top 20
```

```python
for i in top20:
    plt.imshow(faces[i].reshape((64,64)))
    plt.show()
```
